# Exercise 2.1 - PCA for Global Development

```python
import pandas as pd
import seaborn as sns

# Importing Global Development Data
global_development = pd.read_csv("data/worlddev.csv")

# Displaying the newly imported table
display(global_development)
```

## 2.1, Components & Variance
The number of components that one needs depends on the desired cumulative variance and the number of factors that exist originally in the dataset. Here, I decide to go for two component forms - one 2 component PCA, which explains only around 55% of the variance, and one 11 component PCA, which explains around 95% of the variance. This is a notable reduction from the 18 factors that we had before. An example of the variance graph for the unadjusted data shows how inaccurate the method is without any adjustment efforts - we can logically assume that negative variance, and 1 factor explaining almost all of 18, is not correct.

In the following code, I follow three main steps:
- Adjusting Values, where I create new, adjusted dataframes to be used for the PCA analysis.
- PCA Analysis, where I do the PCA analysis graph for variance and create my PCA arrays based on desired n-components.
- Finalizing Dataframes, where I combine the new values with the original string information and give the columns names.

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Adjusting Values
## First, adjusting the dataset so that it only has continuous values for the PCA analysis. Grabs all the indexes that match the selected dtypes from the original dataframe.
gd_continuous = global_development[global_development.select_dtypes(include=["float64", "int64"]).columns]
gd_string = global_development[global_development.select_dtypes(include=["object"]).columns]

## Filling all NaN values with 0, or else the PCA analysis will fail
gd_continuous.fillna(0, inplace=True)

## Standardizing the continuous values for the machine learning algorithm. Gets changed back to array so setting to pandas for visualization (array works for PCA).
gd_continuous_sdr = pd.DataFrame(StandardScaler().fit_transform(gd_continuous))

## Displaying change
display(gd_continuous_sdr)
```

```python
# PCA Analysis
## PCA analysis equation. Does full n PCA components fitted to the continuous data. 
pca = PCA().fit(gd_continuous_sdr)

## Testing prints
# print(pca.shape)
# print(pca.components_.shape)
# print(gd_continuous_sdr.shape)
# display(pd.DataFrame(pca.components_))

## Makes a plot of the sum of the explained variance of each new factor to understand total explained variance. In total, there are 18 factors in the global dev dataframe.
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Cumulative Variance vs Number of Components, Adjusted Data")
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.xticks(np.arange(3, 18, 3)) 
plt.show()

## Exploring what the variance explanation would be if values aren"t adjusted
pca = PCA().fit(gd_continuous)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Cumulative Variance vs Number of Components, Unadjusted Data")
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.xticks(np.arange(3, 18, 3))
plt.show()

## 2/18 Component PCA. Explains ~55% of the variance. Transform added to change the pca to dimensionality reduction.
pca_gd_2 = PCA(n_components=2).fit(gd_continuous_sdr).transform(gd_continuous_sdr)

## 11/18 Component PCA. Explains ~95% of the variance.
pca_gd_11 = PCA(n_components=11).fit(gd_continuous_sdr).transform(gd_continuous_sdr)

## Showing change
print(f"Original Shape: {gd_continuous_sdr.shape}")
print(f"2 Factor PCA Shape: {pca_gd_2.shape}")
```

```python
# Finalizing Dataframes
## Getting number of columns and column names
num_columns = pca_gd_11.shape[1]
column_names = [f"principal component {i+1}" for i in range(num_columns)]

## Establishing dataframes from PCA data
gd_pca_2 = pd.DataFrame(data = pca_gd_2, columns = column_names[0:2])
gd_pca_11 = pd.DataFrame(data = pca_gd_11, columns = column_names)

## recombining with string data
gd_pca_2 = pd.concat([gd_pca_2, gd_string], axis = 1)
gd_pca_11 = pd.concat([gd_pca_11, gd_string], axis = 1)

## Showing change
display(gd_pca_11)
```

## 2.1, Visualizations
Here, I preformed four visualizations:

- The 1st looks at a scatterplot of the 2 factor analysis with a hue with continent. We can"t see very clear patterns, though Africa does appear to concentrate in the lower ranges of principal component 1. Component 2 seems to follow some level of population/land size/gdp, judging by the extraneous variables.
- The 2nd shows a much clearer pattern with income_class, where there is a trend from low -> middle low -> middle high -> high. This implies that a major part of the variance between countries is tied to income and related metrics.
- The 3rd shows 2nd visualization through a pairplot. We can see the disparity in principal component 1 with income class, while principal component 2 appears largely unaffected.
- The 4th shows the 11 factor result as a pairplot. It is more difficult to easily discern effects with the 11 factor, but it does significantly reduce the original number of factors.

```python
# Visualizations
## Plotting the 2 factor pca analysis scatterplot using continent as a hue.
sns.scatterplot(data=gd_pca_2, x="principal component 1", y="principal component 2", hue="continent")
plt.title("Continent vs 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

## Plotting the 2 factor pca analysis scatterplot using income_class as a hue.
sns.scatterplot(data=gd_pca_2, x="principal component 1", y="principal component 2", hue="income_class")
plt.title("Income Class vs 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

## Plotting the 2 factor pca analysis pairplot using income_class as a hue. Changes the histograms to a layered kernel density estimate when hue is used, which calculates the density of features based on surrounding information.
sns.pairplot(gd_pca_2, hue="income_class")
plt.show()

## Plotting the 11 factor pca analysis pairplot.
sns.pairplot(gd_pca_11)
plt.show()
```

# Exercise 2.2 - PCA for Coffee
Exercise 2.2 largely follows the same patterns of 2.1, and so certain explanations in the markdown overview will not be directly repeated here.

```python
# Importing Global Development Data
coffee_data = pd.read_csv("data/df_arabica_clean.csv")

# Displaying the newly imported table
display(coffee_data)
```

## 2.2, Components & Variance
Here, I decide to go for two component forms - one 2 component PCA, which explains only around 70% of the variance, and one 9 component PCA, which explains around 70% of the variance. Interestingly, even though this dataframe had more components than the global development dataframe (19 vs 18), it required fewer factors to achieve the same variance.


```python
# Adjusting Values
## First, adjusting the dataset so that it only has continuous values for the PCA analysis. Grabs all the indexes that match the selected dtypes from the original dataframe.
cf_continuous = coffee_data[coffee_data.select_dtypes(include=["float64", "int64"]).columns]
cf_string = coffee_data[coffee_data.select_dtypes(include=["object"]).columns]

## Filling all NaN values with 0, or else the PCA analysis will fail
cf_continuous.fillna(0, inplace=True)

## Standardizing the continuous values for the machine learning algorithm. Gets changed back to array so setting to pandas for visualization (array works for PCA).
cf_continuous_sdr = pd.DataFrame(StandardScaler().fit_transform(cf_continuous))

## Displaying change
display(cf_continuous_sdr)
```

```python
# PCA Analysis
## PCA analysis equation. Does full n PCA components fitted to the continuous data. 
pca = PCA().fit(cf_continuous_sdr)

## Makes a plot of the sum of the explained variance of each new factor to understand total explained variance. In total, there are 18 factors in the global dev dataframe.
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Cumulative Variance vs Number of Components, Adjusted Data")
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.xticks(np.arange(3, 18, 3)) 
plt.show()

## 2/19 Component PCA. Explains ~70% of the variance. Transform added to change the pca to dimensionality reduction.
pca_cf_2 = PCA(n_components=2).fit(cf_continuous_sdr).transform(cf_continuous_sdr)

## 9/19 Component PCA. Explains ~95% of the variance.
pca_cf_9 = PCA(n_components=9).fit(cf_continuous_sdr).transform(cf_continuous_sdr)

## Showing change
print(f"Original Shape: {cf_continuous_sdr.shape}")
print(f"2 Factor PCA Shape: {pca_cf_2.shape}")
```

```python
# Finalizing Dataframes
## Getting number of columns and column names
num_columns = pca_cf_9.shape[1]
column_names = [f"principal component {i+1}" for i in range(num_columns)]

## Establishing dataframes from PCA data
cf_pca_2 = pd.DataFrame(data = pca_cf_2, columns = column_names[0:2])
cf_pca_9 = pd.DataFrame(data = pca_cf_9, columns = column_names)

## recombining with string data
cf_pca_2 = pd.concat([cf_pca_2, cf_string], axis = 1)
cf_pca_9 = pd.concat([cf_pca_9, cf_string], axis = 1)

## Showing change
display(cf_pca_9)
```

## 2.2, Visualizations
Here, I did for visualizations again:
- In the 1st, I converted Grading Date to a continuous variable and compared principal component 1 and principal component 2 to it. Interestingly, we can that there are very specific days that the data collects at, and that during these days, it appears that principal component 2 is low.
- In the 2nd, I compared the color to the principal components. Nothing to exciting here - shows that the components have a scattered distribution, and with the number of colors, it is hard to make out if there"s any actual pattern.
- In the 3rd, I used a pairplot with KDE representation. Most of principal component 1 and 2 are connected, but there appears to be a few values that are distributed far from the rest.
- In the 4th, I showed the PCA with all 9 factors on a pairplot.


```python
# Visualizations
## Converting
### Remove the ordinal indicators from the "Grading Date" column. Copying, else dataframe would be directly duplicated/affected.
cf_pca_2_datetime = cf_pca_2.copy()
cf_pca_2_datetime["Grading Datetime"] = cf_pca_2_datetime["Grading Date"].str.replace(r"(?<=\d)(st|nd|rd|th)\b", "", regex=True)

### Convert the "Grading Date" column to datetime format
cf_pca_2_datetime["Grading Datetime"] = pd.to_datetime(cf_pca_2_datetime["Grading Datetime"], format="%B %d, %Y")

### Converting Grading Date to a continuous value
cf_pca_2_datetime["Grading Date Continuous"] = cf_pca_2_datetime["Grading Datetime"].apply(lambda x: x.timestamp())

## Plotting the 2 factor pca analysis scatterplot using principal component 2 as a hue and Grading Date Continuous as the y value.
sns.scatterplot(data=cf_pca_2_datetime, x="principal component 1", y="Grading Date Continuous", hue="principal component 2")
plt.title("Grading Date Continuous vs 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Continuous Datetime")
plt.show()

## Plotting the 2 factor pca analysis scatterplot using color as a hue.
sns.scatterplot(data=cf_pca_2, x="principal component 1", y="principal component 2", hue="Color")
plt.title("Color vs 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

## Plotting the 2 factor pca analysis pairplot. This time. used kde to show where the data concentrates.
sns.pairplot(cf_pca_2, kind="kde")
plt.show()

## Plotting the 9 factor pca analysis pairplot.
sns.pairplot(cf_pca_9)
plt.show()
```
