#### Exercise 1.1 - Global Development Data
In Python, the best way to gather information from a csv and put it into a workable format is through the pandas read_csv function. This, alongside display - which allows for a more rich display in Jupyter notebooks then a normal print function - are done below. Import commands for all relevant functions are also done at the beginning of this first code block.

All analysis (part 4) is done throughout the code.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

## Importing Global Development Data
global_development = pd.read_csv("data/worlddata.csv")
diamonds = pd.read_csv("data/diamonds.csv")

# Displaying the newly imported table
display(global_development)
```

##### Part 1 + Part 2 Descriptors
To find the variable types, I first displayed the pandas dtypes of the columns and then displayed it as a pandas dataframe. That result found that almost all objects, aside from the country name and income class, were considered floats by Python. The either continuous or percentage-based nature of this data makes this result expected. Interestingly, this exploration made me discover that Pandas names all mixed-type or string columns as "object", rather than string64 like I expected.

I also returned to this section to reorganize the income_class variable after realizing that, because of it automatically being applied as an object, the order for the incomes displayed oddly (low -> upper middle -> lower middle -> high income). I changed the income classification to categorical over object to fix this.

To show the descriptors, I then took four different parts of the global development data - gdp, gdp per capita, population, and life expectancy - and put them into a pandas dataframe together. I utilized the command .describe(), which gives descriptors for a pandas column, which will only work if the column is of a correct type (i.e. you will got get a min or max for a column with strings). This made me discover that the dataset has missing data, with some countries not having gdp or gdp per capita information. This is something that, for the future, may need to be noted in any summaries that attempt to relate these two together. From this, GDP, GDPPC and Population's values all show how certain countries are far beyond others (i.e. the 75% mark for GDPPC being 17083, and the std being 18657). Something I also discovered when doing this is that the columns in pandas are reformatted to the e+ display if, upon creation, one part of the data would require it. To make the data more readable, I adjusted all float values to display as a .2f type. There is likely a better way to make all rows have different display types, or to make the larger numbers more readable, then I did right now.

```python
# Variable types
## dtype function, where the dtypes is transformed into a pandas dataframe in order to produce a better display.
gd_dtypes = pd.DataFrame(global_development.dtypes, columns=['dtype'])
display(gd_dtypes)

## Fixing categorical nature of income for future graphs
income_categories_order = ['Low income', 'Lower middle income', 'Upper middle income', 'High income']
global_development['income_category'] = pd.Categorical(global_development['income_class'], categories=income_categories_order, ordered=True)

## Showing the change
gd_dtypes = pd.DataFrame([global_development["income_category"].dtypes, global_development["income_class"].dtypes], ["Income Category", "Income Class"], columns=['dtype'])
display(gd_dtypes)

# Descriptors
## Setting Pandas display format for float64 dtype
pd.options.display.float_format = '{:.2f}'.format

## Dataframe for development descriptors
development_descriptors = pd.DataFrame([global_development["gdp"].describe(), global_development["population"].describe(), global_development["gdp_capita"].describe(), global_development["alcohol_consumption"].describe()])
display(development_descriptors)
```

##### Part 2 Graphs
In the first graph, I decided to compare adult literacy rates vs the income category of the country. Notably, the income class appears strongly correlated with adult literacy. Upper middle income countries only begin appearing when adult literacy is at least 70%, and the high income countries when adult literacy is 90%. Additionally, adult literacy hits its peak at around 95% and then drops off as it approaches 100% literacy. Low income and lower middle income countries, however, regardless of the income category, continue to have more countries in all histogram bins.

I then decided to make a scatter plot comparing population and land area. When first making it, the graph shows clear issues, with the vast differences between the largest countries and the smallest countries - something found when looking at the descriptors from before - making the graph unhelpful. By changing to a log graph, the correlation between population and land area becomes much clearer.

Finally, I decided to compare income category with alcohol consumption. Here, it appears that, the higher the income bracket, the greater the countries alcohol consumption. It is a interesting result - perhaps religiosity's potential correlation with income category may impact this?

```python
## Histogram, Adult Literacy vs Income Class
global_histogram = sns.histplot(x=global_development.adult_literacy, hue=global_development.income_category, kde="true", bins=4, multiple="stack")
global_histogram.set(title="Adult Literacy vs Income Class")
plt.show()

## Scatter Plot, Population vs Land Area, no log
global_no_log = sns.scatterplot(x="population", y="land_area", data=global_development)
global_no_log.set(title="Population vs Land Area, No Log")
plt.show()

## Scatter Plot, Population vs Land Area, log
global_log = sns.scatterplot(x="population", y="land_area", data=global_development)
global_log.set(title="Population vs Land Area, Log")
plt.yscale('log')
plt.xscale('log')
plt.show()

## Box Plot, Income Category vs Alcohol Consumption
ax = sns.boxplot(x='income_category', y='alcohol_consumption', data=global_development)
### Establishing medians that can be displayed on the box plots
medians = global_development.groupby('income_category')['alcohol_consumption'].median()
for party, median_val in medians.items():
    ax.text(party, median_val, f'{median_val}', 
            horizontalalignment='center', color='black', weight='bold')
plt.title('Income Class vs Alcohol Consumption')
plt.xticks(rotation=45) # Rotates the x-axis labels to ensure they are readable/do not blend into one another
plt.show()
```

##### Part 3
I then placed most of the data into a correlation heat map. From this, I discovered that the five strongest correlations were, in order:
- Life Expectancy and Water Access
- Life Expectancy and Adult Literacy
- Life Expectancy and GDP Per Capita
- Air Pollution and GDP Per Capita
- GDP and Population

Here, GDP and Population were only the fifth largest correlation, which, considering their directly corresponding formulaic nature, means that the other four correlations are particularly significant. Water access being the strongest is not a particularly significant finding, where access to a basic necessary of life corresponds with how long people live. These values, of course, all play into each other - adult literacy and water access are decently correlated with gdp per capita.

Interestingly, population and population density also appeared to have the weakest correlation out of all relationships, while the expectation would be that a higher population would likely correspond with a higher population density.

Like the histograms before, alcohol consumption also appears to follow alongside better conditions for the average person.

```python
# Correlation Plot/Heatmap
corr_data = global_development[["gdp", "gdp_capita", "labor_rate", "land_area", "life_expectancy", "adult_literacy", "water_access", "air_pollution", "population_density", "population", "alcohol_consumption", "freedom"]]
sns.heatmap(corr_data.corr(), vmin=-1, vmax=1, annot=True)
```

##### Part 5
Because of the findings in parts 2 and 3, I decided to compare the alcohol consumption below and above the median, alongside generosity, with income category. Here, we find that, aside from low income, higher alcohol consumption in the same income category appears to correlate with lower generosity. This is most prominent in the median for high income, where higher alcohol consumption shows a generosity of <0, while a lower alcohol consumption has a generosity of around 1.5. 

```python
# Violin Plot
global_development['Above Alcohol Consumption Median'] = (global_development['alcohol_consumption'] > global_development['alcohol_consumption'].median())
sns.violinplot(data=global_development, x="generosity", y="income_category", hue="Above Alcohol Consumption Median", split=True, gap=.1)
```

#### Exercise 1.2 - Diamonds Data
For the diamonds dataset, I followed a similar path to the global development dataset. A difference in the setup was that I dropped the extra numbered column, since importing the read_csv normally, due to how the diamonds dataset was constructed, would result in two identical numbered columns.

```python
## Importing Diamonds Data
diamonds = pd.read_csv("data/diamonds.csv")

# Displaying the newly imported table
## Dropping Unnamed: 0 column, since column numbers are already automatically implemented, so it adds nothing to the display
diamonds = diamonds.drop(columns=['Unnamed: 0'])
display(diamonds)
```

##### Part 1 + Part 2 Descriptors
For the categorical variable in the diamonds dataset, I decided to fix the display of the cut categories. The descriptors also show how carat and price have a wide-ranging distribution of information, while depth is much more concentrated (the 25% value is 61, while the 75% value is 62.5, with an std of 1.43).

```python
# Variable types
## dtype function, where the dtypes is transformed into a pandas dataframe in order to produce a better display.
d_dtypes = pd.DataFrame(diamonds.dtypes, columns=['dtype'])
display(d_dtypes)

## Fixing categorical nature of income for future graphs
cut_categories_order = ['Fair', 'Good', 'Very Good', 'Premium', "Ideal"]
diamonds['cut_categories'] = pd.Categorical(diamonds['cut'], categories=cut_categories_order, ordered=True)

## Showing the change
d_dtypes = pd.DataFrame([diamonds["cut"].dtypes, diamonds["cut_categories"].dtypes], ["Cut", "Cut Categories"], columns=['dtype'])
display(d_dtypes)

# Descriptors
## Dataframe for development descriptors
diamond_descriptors = pd.DataFrame([diamonds["carat"].describe(), diamonds["depth"].describe(), diamonds["price"].describe()])
display(diamond_descriptors)
```

##### Part 2 Graphs
First, I decided to see how the diamonds carat compares with its cut category. Notably, ideal diamonds are far more common in the lower carats before falling off, while the remaining classification types generally keep a similar ratio throughout. While carat is normally associated with better diamonds, it is directly related to a diamonds weight. It is likely that only the most perfect diamonds are considered ideal, and that, as carat is likely loosely associated with size, larger carats are more likely to have at least one imperfection.

Next, I made a scatter plot comparing carat to price. We can see a correlation here between a higher carat rating and a higher price on the diamond. Interestingly, it seems that the diamond classifiers don't like to place a diamond at something like 3.99 when instead it could be placed at 4, creating a weak binning of the different diamond carats.

Finally, I made a box plot comparing the diamond depth to the cut categories. It appears that a worse classification does have a correlation with the grading, though with a significant amount of values escaping the box plots standard deviation range.

```python
## Histogram, Carat Weight vs Cut Categories
global_histogram = sns.histplot(x=diamonds.carat, hue=diamonds.cut_categories, kde="true", bins=5, element="step")
global_histogram.set(title="Carat Weight vs Cut Categories")
plt.show()

## Scatter Plot, Carat vs Price, no log
global_no_log = sns.scatterplot(x="carat", y="price", data=diamonds)
global_no_log.set(title="Carat vs Price, No Log")
plt.show()

## Box Plot, Diamond Depth vs Cut Categories
ax = sns.boxplot(x='cut_categories', y='depth', data=diamonds)
### Establishing medians that can be displayed on the box plots
medians = diamonds.groupby('cut_categories')['depth'].median()
for party, median_val in medians.items():
    ax.text(party, median_val, f'{median_val}', 
            horizontalalignment='center', color='black', weight='bold')
plt.title('Diamond Depth vs Cut Categories')
plt.xticks(rotation=45) # Rotates the x-axis labels to ensure they are readable/do not blend into one another
plt.show()
```

##### Part 3
The correlation heat map between the diamond values discovered a few things - that carat, price, x, y, and z are all strongly correlated with one another, and only have a weak association with depth and table. Depth and table, on the other hand, have their strongest relationship with one another, yet it is weaker then the almost perfect relationship that the other values hold.

```python
# Correlation Plot/Heatmap
corr_data = diamonds[["carat", "table", "depth", "price", "x", "y", "z"]]
sns.heatmap(corr_data.corr(), vmin=-1, vmax=1, annot=True)
```

##### Part 5
Finally, I decided to compare the price medians and color grading to the carat. Across the board, one can see that a higher price median is associated with a higher carat. However, the colors I, J, and H appear to be generally higher priced then the other colors.

```python
# Violin Plot
diamonds['price_medians'] = (diamonds['price'] > diamonds['price'].median())
sns.violinplot(data=diamonds, x="carat", y="color", hue="price_medians", split=True, gap=.1, inner="quart")
```
