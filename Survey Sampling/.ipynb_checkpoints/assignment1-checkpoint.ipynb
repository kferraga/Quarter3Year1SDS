## #1, 3, & 4: Importing, Labeling, and Describing Province17
When importing the province17 txt file, I set the delimiter to be a tab ("\t") so that it creates multiple columns and changed the encoding to Latin-1 due the encoding of the txt file (Latin-1 encodes for letters in the Latin alphabet, UTF-8 is for all valid character code points).

I then moved on to relabeling my values to be more readable, like changing POP17 to Population in 2017. I also changed the Urbanization value names using a new function "value_mapping" so it now noted if the province had less than or more than 50% urbanization.

With this information, I moved onto making some descriptors for the dataset. First, I created a new variable for the % of unemployment instead of an absolute value. I then moved onto a histogram of said %, with a kde line, which visualizes how the histogram is distributed. Notably, there is a slight left skew, with unemployment's peak being around 4.5% and ranging from 3.5% to 6.5%. I finally decided to make a box plot with unemployment % with urbanization, noting that the median value for the less-urbanized regions (4.54%) was less than those with higher levels of urbanization (5.01%).

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

def value_mapping (pdata, variable_name, map_dict, new_variable = None):
    """Given a pandas dataframe, a variable name, and a desired replacement dictionary, change how the variables values are recorded. Can also create a new variable instead of replacing by setting a value for new_variable."""
    temp_value = [map_dict.get(value, value) for value in pdata[variable_name]]
    if new_variable:
        pdata[new_variable] = temp_value
    else:
        pdata[variable_name] = temp_value
    return pdata
    
    
## Importing Province17 Data
province17 = pd.read_csv("files/assignment1/province17.txt", delimiter = '\t', encoding='latin-1')

# Displaying the newly imported table (done later as well)
display(province17)

## Labeling
# Creates a dictionary of the label names
label_mapping = {
    "id": "ID",
    "POP17": "Population in 2017",
    "LAB17": "Labor Force in 2017",
    "UE17": "Unemployed in 2017",
    "HOU17": "Households in 2017",
    "URB17_new": "Urbanization in 2017"
}
# Adding labels to each of the column names
province17.rename(columns=label_mapping, inplace=True)

# Renaming urbanization's values with the new function
province17 = value_mapping(province17, "Urbanization in 2017", {0: "Less Than 50%", 1: "Greater Than 50%"})


## Descriptors
# Adding new variable. Rounded to ensure that it is only 2 decimal places.
province17["Unemployed % in 2017"] = round(province17["Unemployed in 2017"]/province17["Population in 2017"]*100, 2)
display(province17)

# Histogram of Unemployed %, including kde line
sns.histplot(province17["Unemployed % in 2017"], kde = True)
plt.show()

## Box Plot, including the addition of median values in text
ax = sns.boxplot(x='Urbanization in 2017', y='Unemployed % in 2017', data=province17)
medians = round(province17.groupby('Urbanization in 2017')['Unemployed % in 2017'].median(), 2)
for party, median_val in medians.items():
    ax.text(party, median_val, f'{median_val}%', 
            horizontalalignment='center', color='white')
plt.title('Urbanization vs Unemployed % in 2017, Finland')
plt.show()
```
